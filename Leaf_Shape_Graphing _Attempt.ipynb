# This small section of code depicts a failed attempt at plotting out the shape of a theoretical leaf.

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/DanChitwood/grapevine_shoots/master/Datasets/Procrustes_all.csv'

df = pd.read_csv(url)

raw_data=pd.DataFrame(df).to_numpy()

raw_data_clean=np.delete(raw_data,[range(0,50)],1)
```

```python
pca = PCA(n_components=2)

PCs = pca.fit_transform(raw_data_clean)

inverse_PCA = pca.inverse_transform(PCs)
```

```python
#This plotting line simply plots all 21 points from the output of the inverse transform function in order
plt.plot(inverse_PCA[0,0::2], inverse_PCA[0,1::2])
```

As you may be able to tell, this simplified plotting method correctly plots part of the edge of the leaf, but the veins and area near the stem of the leaf are a confusing mess.  The function used in final code plots the leaves in the corrrect order.
